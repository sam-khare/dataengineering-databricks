```python
#ISO Date Input format - 2024-10-27T13:00:00 (yyyy-MM-ddTHH:mm:ss)
#Read data of Input and get the day of the week - Like Sun,Mon etc
#create Input_date widget
dbutils.widgets.text("input_date","")
```

```python
#read the value from widget
input_date = dbutils.widgets.get("input_date")
print(input_date)
```

```python
# Format the input_date to extract the day of the week
#set conf to legacy to support datatime pattern
spark.conf.set("spark.sql.legacy.timeParserPolicy","LEGACY")
#from pyspark.sql.function import date_format, to_timestamp
input_day = spark.sql(f"""
                      select date_format(to_timestamp('{input_date}',"yyyy-MM-dd'T'HH:mm:ss"),'E')
                      """
                      ).collect()[0][0]
#collect since we need output in python
print(input_day)
```

```python
#export the day of week using taskvalues to use in next task
dbutils.jobs.taskValues.set("input_day",input_day)
```
